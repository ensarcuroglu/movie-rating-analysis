```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/tmdb_5000_movies.csv")
```

```python
# Tarihi datetime formatına çevir
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
df["release_year"] = df["release_date"].dt.year
```

```python
yearly_avg = df.groupby("release_year")["vote_average"].mean()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=yearly_avg.index, y=yearly_avg.values, marker="o")
plt.title("Yıllara Göre Ortalama Film Puanları")
plt.xlabel("Yıl")
plt.ylabel("Ortalama Puan")
plt.grid(True)
plt.tight_layout()
plt.savefig("../visuals/yearly_vote_average.png", dpi=300)
plt.show()
```
