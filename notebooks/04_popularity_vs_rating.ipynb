```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/tmdb_5000_movies.csv")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="popularity", y="vote_average", alpha=0.5)
plt.xlabel("Poplerlik Skoru")
plt.ylabel("Ortalama Puan")
plt.title("Poplerlik ile Puan Arasndaki liki")
plt.tight_layout()
plt.savefig("../visuals/popularity_vs_rating.png", dpi=300)
plt.show()
```

### 10.  Poplerlik ile Puan Arasndaki liki
Bu grafik, TMDBnin poplerlik skoru ile izleyici puan arasnda gl bir korelasyon olmadn gstermektedir.
Baz filmler yksek poplerlie ramen ortalama puan alrken, baz dk poplerlikteki filmler izleyici tarafndan olduka beenilmitir.
Bu durum, poplerliin genellikle pazarlama etkisi, byk stdyo btesi veya oyuncu kadrosu gibi faktrlerle olutuunu; fakat bu faktrlerin izleyici beenisiyle dorudan rtmeyebileceini ortaya koyar.

### 10.  Popularity vs Rating Relationship
This chart shows that there is no strong correlation between TMDB's popularity score and viewer rating. Some films have high popularity but receive average ratings, while some low-popularity films are highly appreciated by audiences. This reveals that popularity often stems from factors like marketing, big studio budgets, or cast, which may not directly align with viewer satisfaction.
