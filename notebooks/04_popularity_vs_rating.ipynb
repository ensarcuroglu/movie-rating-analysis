```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/tmdb_5000_movies.csv")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="popularity", y="vote_average", alpha=0.5)
plt.xlabel("Popülerlik Skoru")
plt.ylabel("Ortalama Puan")
plt.title("Popülerlik ile Puan Arasındaki İlişki")
plt.tight_layout()
plt.savefig("../visuals/popularity_vs_rating.png", dpi=300)
plt.show()
```

### 10. 🔥 Popülerlik ile Puan Arasındaki İlişki
Bu grafik, TMDB’nin popülerlik skoru ile izleyici puanı arasında güçlü bir korelasyon olmadığını göstermektedir.
Bazı filmler yüksek popülerliğe rağmen ortalama puan alırken, bazı düşük popülerlikteki filmler izleyici tarafından oldukça beğenilmiştir.
Bu durum, popülerliğin genellikle pazarlama etkisi, büyük stüdyo bütçesi veya oyuncu kadrosu gibi faktörlerle oluştuğunu; fakat bu faktörlerin izleyici beğenisiyle doğrudan örtüşmeyebileceğini ortaya koyar.

### 10. 🔥 Popularity vs Rating Relationship
This chart shows that there is no strong correlation between TMDB's popularity score and viewer rating. Some films have high popularity but receive average ratings, while some low-popularity films are highly appreciated by audiences. This reveals that popularity often stems from factors like marketing, big studio budgets, or cast, which may not directly align with viewer satisfaction.
