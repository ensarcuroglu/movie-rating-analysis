```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/tmdb_5000_movies.csv")
```

```python
# Sıfır bütçeli filmleri çıkar
df_budget = df[df["budget"] > 0]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_budget, x="budget", y="vote_average", alpha=0.5)
plt.xlabel("Bütçe")
plt.ylabel("Ortalama Puan")
plt.title("Bütçe ile Ortalama Puan Arasındaki İlişki")
plt.xscale("log")  # Logaritmik ölçek büyük değerleri daha okunabilir kılar
plt.tight_layout()

# 📁 visuals klasörüne kayıt
plt.savefig("../visuals/budget_vs_rating.png", dpi=300)

plt.show()
```
