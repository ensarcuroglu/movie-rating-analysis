```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/tmdb_5000_movies.csv")
```

```python
popular_df = df[df["vote_count"] > 1000]
```

```python
top_10 = popular_df.sort_values("vote_average", ascending=False).head(10)
top_10[["title", "vote_average", "vote_count"]]
```

```python
bottom_10 = popular_df.sort_values("vote_average", ascending=True).head(10)
bottom_10[["title", "vote_average", "vote_count"]]
```

```python
plt.figure(figsize=(14,6))

# En yksek puanl
plt.subplot(1,2,1)
sns.barplot(x=top_10["vote_average"], y=top_10["title"], palette="Greens_r")
plt.title("En Beenilen 10 Film (1000+ Oy)")
plt.xlabel("Puan")
plt.ylabel("")

# En dk puanl
plt.subplot(1,2,2)
sns.barplot(x=bottom_10["vote_average"], y=bottom_10["title"], palette="Reds")
plt.title("En Beenilmeyen 10 Film (1000+ Oy)")
plt.xlabel("Puan")
plt.ylabel("")

plt.tight_layout()
plt.savefig("../visuals/best_and_worst_movies.png", dpi=300)
plt.show()
```
